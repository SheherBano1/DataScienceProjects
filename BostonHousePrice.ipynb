"Linear Regression model implementation without buit-in liabraries."

```python
import numpy as np

# Generate synthetic data
np.random.seed(2)
X = np.random.rand(100, 1) * 10  # 100 data points, single feature
Y = 5 * X + np.random.randn(100, 1) * 5  # Linear relationship with some noise

# Split the data manually into training and testing sets (80% train, 20% test)
split_index = int(0.8 * len(X))
X_train, X_test = X[:split_index], X[split_index:]
Y_train, Y_test = Y[:split_index], Y[split_index:]

# Implement Linear Regression from scratch
class LinearRegression:
    def __init__(self):
        self.m = None  # Slope
        self.c = None  # Intercept
    
    def fit(self, X, Y):
        # Calculate slope (m) and intercept (c) using the formulas:
        # m = ((X - mean_X) * (Y - mean_Y)) / ((X - mean_X)^2)
        # c = mean_Y - m * mean_X
        X_mean = np.mean(X)
        Y_mean = np.mean(Y)
        
        numerator = np.sum((X - X_mean) * (Y - Y_mean))
        denominator = np.sum((X - X_mean) ** 2)
        
        self.m = numerator / denominator
        self.c = Y_mean - self.m * X_mean
    
    def predict(self, X):
        # Prediction: Y = m * X + c
        return self.m * X + self.c
    
    def evaluate(self, X, Y):
        # Calculate Mean Squared Error (MSE) as a metric
        predictions = self.predict(X)
        mse = np.mean((Y - predictions) ** 2)
        return mse

# Initialize the model
model = LinearRegression()

# Train the model
model.fit(X_train, Y_train)

# Test the model
predictions = model.predict(X_test)

# Evaluate the model
mse = model.evaluate(X_test, Y_test)

# Display the results
print(f"Model Coefficients: Slope (m) = {model.m}, Intercept (c) = {model.c}")
print(f"Mean Squared Error on Test Set: {mse}")

# Visualize results
import matplotlib.pyplot as plt

plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression from Scratch')
plt.legend()
plt.show()

```

```python
import numpy as np

# Generate synthetic data
np.random.seed(2)
X = np.random.rand(100, 1) * 10  # 100 data points, single feature
Y = 5 * X + np.random.randn(100, 1) * 5  # Linear relationship with some noise

# Split data manually into training and testing sets (80% train, 20% test)
split_index = int(0.8 * len(X))
X_train, X_test = X[:split_index], X[split_index:]
Y_train, Y_test = Y[:split_index], Y[split_index:]

# Decision Tree implementation
class DecisionTree:
    def __init__(self, max_depth=5, min_samples_split=2):
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.tree = None

    def fit(self, X, Y):
        self.tree = self._build_tree(X, Y, depth=0)

    def _build_tree(self, X, Y, depth):
        # Stop splitting if depth exceeds max_depth or samples are too few
        if depth >= self.max_depth or len(X) < self.min_samples_split:
            return {'value': np.mean(Y)}

        # Find the best split
        best_split = self._get_best_split(X, Y)
        if not best_split:
            return {'value': np.mean(Y)}

        left_tree = self._build_tree(best_split['X_left'], best_split['Y_left'], depth + 1)
        right_tree = self._build_tree(best_split['X_right'], best_split['Y_right'], depth + 1)

        return {
            'feature_index': best_split['feature_index'],
            'threshold': best_split['threshold'],
            'left': left_tree,
            'right': right_tree,
        }

    def _get_best_split(self, X, Y):
        best_split = {}
        best_mse = float('inf')

        for feature_index in range(X.shape[1]):
            thresholds = np.unique(X[:, feature_index])
            for threshold in thresholds:
                X_left, Y_left, X_right, Y_right = self._split(X, Y, feature_index, threshold)

                if len(X_left) == 0 or len(X_right) == 0:
                    continue

                mse = self._calculate_mse(Y_left, Y_right)
                if mse < best_mse:
                    best_mse = mse
                    best_split = {
                        'feature_index': feature_index,
                        'threshold': threshold,
                        'X_left': X_left,
                        'Y_left': Y_left,
                        'X_right': X_right,
                        'Y_right': Y_right,
                    }

        return best_split if best_split else None

    def _split(self, X, Y, feature_index, threshold):
        left_indices = X[:, feature_index] <= threshold
        right_indices = ~left_indices
        return X[left_indices], Y[left_indices], X[right_indices], Y[right_indices]

    def _calculate_mse(self, Y_left, Y_right):
        mse_left = np.var(Y_left) * len(Y_left)
        mse_right = np.var(Y_right) * len(Y_right)
        return (mse_left + mse_right) / (len(Y_left) + len(Y_right))

    def predict(self, X):
        predictions = np.array([self._predict_single(x, self.tree) for x in X])
        return predictions

    def _predict_single(self, x, tree):
        if 'value' in tree:
            return tree['value']

        feature_index = tree['feature_index']
        if x[feature_index] <= tree['threshold']:
            return self._predict_single(x, tree['left'])
        else:
            return self._predict_single(x, tree['right'])

# Random Forest implementation
class RandomForest:
    def __init__(self, n_estimators=10, max_depth=5, min_samples_split=2, max_features=None):
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.max_features = max_features
        self.trees = []

    def fit(self, X, Y):
        self.trees = []
        n_features = X.shape[1]
        self.max_features = self.max_features or n_features

        for _ in range(self.n_estimators):
            # Bootstrap sampling
            indices = np.random.choice(len(X), len(X), replace=True)
            X_sample, Y_sample = X[indices], Y[indices]

            # Build tree
            tree = DecisionTree(max_depth=self.max_depth, min_samples_split=self.min_samples_split)
            tree.fit(X_sample, Y_sample)
            self.trees.append(tree)

    def predict(self, X):
        # Aggregate predictions from all trees
        predictions = np.array([tree.predict(X) for tree in self.trees])
        return np.mean(predictions, axis=0)

    def evaluate(self, X, Y):
        predictions = self.predict(X)
        mse = np.mean((Y - predictions) ** 2)
        return mse

# Train and evaluate Random Forest model
model = RandomForest(n_estimators=10, max_depth=5, min_samples_split=2)
model.fit(X_train, Y_train)
predictions = model.predict(X_test)
mse = model.evaluate(X_test, Y_test)

print(f"Mean Squared Error on Test Set: {mse}")

# Optional: Visualize results
import matplotlib.pyplot as plt

plt.scatter(X, Y, color='blue', label='Data Points')
plt.scatter(X_test, predictions, color='red', label='Predictions')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Random Forest from Scratch')
plt.legend()
plt.show()
```

```python
import numpy as np

# Generate synthetic data
np.random.seed(2)
X = np.random.rand(100, 1) * 10  # 100 data points, single feature
Y = 5 * X + np.random.randn(100, 1) * 5  # Linear relationship with some noise

# Split data manually into training and testing sets (80% train, 20% test)
split_index = int(0.8 * len(X))
X_train, X_test = X[:split_index], X[split_index:]
Y_train, Y_test = Y[:split_index], Y[split_index:]

# Decision Tree for XGBoost
class DecisionTree:
    def __init__(self, max_depth=3, min_samples_split=2, gamma=0):
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.gamma = gamma
        self.tree = None

    def fit(self, X, gradients, hessians):
        self.tree = self._build_tree(X, gradients, hessians, depth=0)

    def _build_tree(self, X, gradients, hessians, depth):
        if depth >= self.max_depth or len(X) < self.min_samples_split:
            return {'value': -np.sum(gradients) / (np.sum(hessians) + 1e-8)}

        best_split = self._get_best_split(X, gradients, hessians)
        if not best_split:
            return {'value': -np.sum(gradients) / (np.sum(hessians) + 1e-8)}

        left_tree = self._build_tree(
            best_split['X_left'], best_split['grad_left'], best_split['hess_left'], depth + 1
        )
        right_tree = self._build_tree(
            best_split['X_right'], best_split['grad_right'], best_split['hess_right'], depth + 1
        )

        return {
            'feature_index': best_split['feature_index'],
            'threshold': best_split['threshold'],
            'left': left_tree,
            'right': right_tree,
        }

    def _get_best_split(self, X, gradients, hessians):
        best_gain = -float('inf')
        best_split = None

        for feature_index in range(X.shape[1]):
            thresholds = np.unique(X[:, feature_index])
            for threshold in thresholds:
                X_left, X_right, grad_left, grad_right, hess_left, hess_right = self._split(
                    X, gradients, hessians, feature_index, threshold
                )

                if len(X_left) == 0 or len(X_right) == 0:
                    continue

                gain = self._calculate_gain(grad_left, grad_right, hess_left, hess_right)
                if gain > best_gain:
                    best_gain = gain
                    best_split = {
                        'feature_index': feature_index,
                        'threshold': threshold,
                        'X_left': X_left,
                        'X_right': X_right,
                        'grad_left': grad_left,
                        'grad_right': grad_right,
                        'hess_left': hess_left,
                        'hess_right': hess_right,
                    }

        return best_split if best_gain > self.gamma else None

    def _split(self, X, gradients, hessians, feature_index, threshold):
        left_indices = X[:, feature_index] <= threshold
        right_indices = ~left_indices
        return (
            X[left_indices], X[right_indices],
            gradients[left_indices], gradients[right_indices],
            hessians[left_indices], hessians[right_indices],
        )

    def _calculate_gain(self, grad_left, grad_right, hess_left, hess_right):
        gain = (
            (np.sum(grad_left) ** 2 / (np.sum(hess_left) + 1e-8)) +
            (np.sum(grad_right) ** 2 / (np.sum(hess_right) + 1e-8)) -
            ((np.sum(grad_left) + np.sum(grad_right)) ** 2 / (np.sum(hess_left) + np.sum(hess_right) + 1e-8))
        )
        return gain

    def predict(self, X):
        return np.array([self._predict_single(x, self.tree) for x in X])

    def _predict_single(self, x, tree):
        if 'value' in tree:
            return tree['value']

        feature_index = tree['feature_index']
        if x[feature_index] <= tree['threshold']:
            return self._predict_single(x, tree['left'])
        else:
            return self._predict_single(x, tree['right'])





  
```

```python
class XGBoost:
    def __init__(self, n_estimators=10, max_depth=3, learning_rate=0.1, min_samples_split=2, gamma=0):
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.learning_rate = learning_rate
        self.min_samples_split = min_samples_split
        self.gamma = gamma
        self.trees = []

    def fit(self, X, Y):
        self.trees = []
        predictions = np.zeros_like(Y.ravel())

        for _ in range(self.n_estimators):
            # Compute gradients and hessians
            gradients = 2 * (predictions - Y.ravel())  # Gradient of MSE
            hessians = np.ones_like(Y.ravel()) * 2     # Hessian is constant for MSE

            # Train a decision tree on the gradients and hessians
            tree = DecisionTree(
                max_depth=self.max_depth,
                min_samples_split=self.min_samples_split,
                gamma=self.gamma
            )
            tree.fit(X, gradients, hessians)
            self.trees.append(tree)

            # Update predictions
            predictions += self.learning_rate * tree.predict(X)

    def predict(self, X):
        # Initialize predictions as zero
        predictions = np.zeros(X.shape[0])

        # Add contributions from all trees
        for tree in self.trees:
            predictions += self.learning_rate * tree.predict(X)

        return predictions

    def evaluate(self, X, Y):
        predictions = self.predict(X)
        mse = np.mean((Y.ravel() - predictions) ** 2)
        return mse
```

```python
# Train the model
model = XGBoost(n_estimators=10, max_depth=3, learning_rate=0.1, gamma=0)
model.fit(X_train, Y_train)

# Predict and evaluate
predictions = model.predict(X_test)
mse = model.evaluate(X_test, Y_test)

print(f"Mean Squared Error on Test Set: {mse}")
```

Comparing the models using RMSE and R square matrices.

```python
import numpy as np

def rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred) ** 2))

def r2_score(y_true, y_pred):
    ss_res = np.sum((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    return 1 - (ss_res / ss_tot)
```

```python
rmse_lr = rmse(Y_test,mse)
r2_lr = r2_score(Y_test,mse)
rmse_xgb = rmse(Y_test,mse)
r2_xgb = r2_score(Y_test,mse)
rmse_rf = rmse(Y_test,mse)
r2_rf = r2_score(Y_test,mse)

```

```python
print("Model Performance:")
print(f"Linear Regression: RMSE = {rmse_lr:.4f}, R = {r2_lr:.4f}")
print(f"Random Forest:     RMSE = {rmse_rf:.4f}, R = {r2_rf:.4f}")
print(f"XGBoost:           RMSE = {rmse_xgb:.4f}, R = {r2_xgb:.4f}")
```
