```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_style("whitegrid")

# Summary statistics
summary_stats = df.describe()

# Plot distributions of key features
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
features = ["age", "trestbps", "chol", "thalach", "oldpeak", "ca", "slope", "thal", "cp"]

for ax, feature in zip(axes.flatten(), features):
    sns.histplot(df[feature], bins=20, kde=True, ax=ax)
    ax.set_title(f"Distribution of {feature}")

plt.tight_layout()
plt.show()

summary_stats
```

```python
# Compute correlation matrix
correlation_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()

# Box plots of key features against target variable
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
features_to_plot = ["age", "trestbps", "chol", "thalach", "oldpeak", "ca"]

for ax, feature in zip(axes.flatten(), features_to_plot):
    sns.boxplot(x="target", y=feature, data=df, ax=ax, palette="Set2")
    ax.set_title(f"{feature} vs. Target")

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X=df.drop(columns='target', axis=1)
Y=df['target']
```

```python
X
```

```python
Y
```

#As we have divided our dataset into 'X' and 'Y' now we will pass into train,test and split.

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2, random_state=2, stratify=Y)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction=model.predict(X_train)
training_data_acc=accuracy_score(X_train_prediction,Y_train)
```

```python
training_data_acc
```

#So our training data accuracy is 85%

```python
X_train_prediction=model.predict(X_test)
test_data_acc=accuracy_score(X_train_prediction,Y_test)
```

```python
test_data_acc
```
