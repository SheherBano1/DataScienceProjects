```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df =pd.read_csv('Titanic Dataset.csv')
df.head(20)
```

```python
x = df["Age"].mean()  # Calculate the mean
df["Age"] = df["Age"].replace("", np.nan)  # Replace empty strings with NaN
df["Age"] = df["Age"].fillna(x) # Fill NaN values with the mean
df["Age"]=df["Age"].astype("int")#Change the datatype of age from float to int
df.rename(columns={"Sex": "Gender"}, inplace=True)#Changed the column name
df.head(30)


```

```python
df["Cabin"]=df["Cabin"].replace("",np.nan) # Replace empty strings with NaN
df["Cabin"]=df["Cabin"].fillna("C123")#Fill NaN values with C123
df.head(30)
```

```python
df["Embarked"]=df["Embarked"].replace(" ",np.nan)# Replace empty strings with NaN
df["Embarked"]=df["Embarked"].fillna("Q")#Fill NaN values with "Q"
```

```python
# Step 1: Group and Calculate Percentages
grouped = df.groupby(["Gender", "Survived"]).size().reset_index(name="Count")
total_counts = grouped.groupby("Gender")["Count"].transform("sum")
grouped["Percentage"] = (grouped["Count"] / total_counts) * 100

# Step 2: Pivot the Data for Stacked Bar Plot
pivot_data = grouped.pivot(index="Gender", columns="Survived", values="Percentage").fillna(0)

# Step 3: Plot Stacked Bar Chart
fig, ax = plt.subplots(figsize=(8, 6))
bar_chart = pivot_data.plot(
    kind="bar", stacked=True, figsize=(8, 6), colormap="coolwarm", ax=ax
)

# Step 4: Add Percentage Labels
for gender_idx, gender in enumerate(pivot_data.index):
    bottom_offset = 0  # Tracks the starting position for stacking
    for survived in pivot_data.columns:
        value = pivot_data.loc[gender, survived]
        if value > 0:  # Add label only for non-zero percentages
            ax.text(
                gender_idx,  # X-coordinate
                bottom_offset + value / 2,  # Y-coordinate (center of the bar section)
                f"{value:.1f}%",  # Text label
                ha="center", va="center", color="black", fontsize=10
            )
            bottom_offset += value  # Increment the bottom offset for the next stack

# Step 5: Customize Plot
plt.title("Survival Percentage by Gender")
plt.ylabel("Percentage")
plt.xlabel("Gender")
plt.legend(title="Survived", labels=["No", "Yes"], loc="upper right")
plt.tight_layout()
plt.show()
```

#From the above barchart we come to know that more female have survived in the accident of Titanic than males

```python
numeric_features = df.select_dtypes(include=["number"])

# Step 2: Define a function to detect outliers using the IQR method
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)  # First quartile (25th percentile)
    Q3 = data[column].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1  # Interquartile Range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Step 3: Detect and handle outliers for each numeric column
for column in numeric_features.columns:
    print(f"\nAnalyzing column: {column}")
    outliers, lower, upper = detect_outliers_iqr(df, column)
    print(f"Outliers detected: {len(outliers)}")
    print(f"Lower bound: {lower}, Upper bound: {upper}")
    
    # Visualize with a boxplot
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[column])
    plt.title(f"Boxplot for {column} (Before Outlier Handling)")
    plt.show()
    
    # Handle outliers by capping values
    df[column] = np.where(df[column] < lower, lower, df[column])  # Cap lower outliers
    df[column] = np.where(df[column] > upper, upper, df[column])  # Cap upper outliers
    
    # Re-visualize after capping
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[column])
    plt.title(f"Boxplot for {column} (After Outlier Handling)")
    plt.show()
```

#The above code identifies outliers in numeric columns of the Titanic dataset using the IQR (Interquartile Range) method. It calculates each column's lower and upper bounds, detects outliers outside these bounds, and visualizes the data using boxplots. Outliers are then managed by capping values that fall outside the bounds of the respective thresholds. The process includes before and after visualizations of boxplots for each column.
