#Extractive text summarization

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc=nlp(text)
```

```python
tokens = [token.text.lower() for token in doc
          if not token.is_punct and 
          token.text !='\n']
```

```python
tokens
```

```python
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation

tokens1 = []
stopwords = list(STOP_WORDS)
allowed_pos = ['ADJ', 'PROPN', 'VERB', 'NOUN']

for token in doc:
    if token.text in stopwords or token.text in punctuation:
        continue  # Skip stopwords and punctuation

    if token.pos_ in allowed_pos:  # Check if token's POS is in the allowed list
        tokens1.append(token.text)

print(tokens1)  # Print filtered tokens

```

```python
from collections import Counter
```

```python
word_freq=Counter(tokens) #Here we created a variable in which we store frequency of our words
```

```python
word_freq
```

```python
max_freq=max(word_freq.values())
```

```python
max_freq
```

```python
#Now we are normalizing our values
for word in word_freq.keys():
    word_freq[word]=word_freq[word]/max_freq
```

```python
word_freq
```

#Here we can notice that the more freqent words have the freqency of 1 while other words have lower values

```python
sent_token=[sent.text for sent in doc.sents]
```

```python
sent_token
```

```python
sent_score={}
for sent in sent_token:
    for word in sent.split():#splits each sentene into words
        if word.lower() in word_freq.keys():#checks if lowercase words exist in the word dictionary
            if sent not in sent_score.keys():
                sent_score[sent]= word_freq[word]
            else:
                sent_score[sent] +=word_freq[word]
                
        print(word)
```

```python
sent_score
```

```python
#Calculating score of sentences
import pandas as pd
pd.DataFrame(list(sent_score.items()),columns=['Sentence','Score'])
```

```python
from heapq import nlargest
```

```python
num_sentences=3
nlargest(num_sentences,sent_score,key=sent_score.get)
```
